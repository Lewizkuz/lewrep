Monitasoisen indeksin riviin viitataan joko 1-tason riviotsikolla tai monikolla, jossa on useamman tason otsikot järjestyksessä.

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta4/esim1.txt', index_col=[1,0])
                
print(df)


print('\n------------------\n')

print(df.loc['Jokerit'])
    
print('\n------------------\n')

print(df.loc[('Jokerit',2011)])

print('\n------------------\n')

# siivutusta voi myös käyttää, mutta sitä varten DataFrame pitää olla lajiteltu
df.sort_index(inplace=True)

print(df.loc['JYP':'Jukurit'])
```

## DataFramen yhdistämien
DataFrameja (ja Seriesejä) voi yhdistää muutamalla tavalla:
<a id="3"></a>
### pandas.merge
`pandas.merge`-funktio yhdistää DataFramet avaimen perusteella samaan tapaan kuin relaatiotietokantojen `join`-toiminto.

Jos funktiolle ei kerrota, minkä kentän mukaan yhdistetään, `merge` käyttää päällekkäisiä (samoja) sarakenimiä "yhdistämiskenttänä". Fiksumpaa on kuitenkin määrittää yhdistämiseen käytettävä kenttä `on`-paremetrilla.

```python
df1 = pd.DataFrame({'avain': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'avain': ['a', 'b', 'd'],
                    'data2': range(3)})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

df3 = pd.merge(df1, df2)

# parempi: df3 = pd.merge(df1, df2, on = 'avain')

print(df3)
```

Jos yhdistämiseen käytettävien sarakkeiden nimet eroavat DataFrameissa, voidaan ne määrittää erikseen `left_on`- ja `right_on`-parametreilla:

```python
df1 = pd.DataFrame({'avain1': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'avain2': ['a', 'b', 'd'],
                    'data2': range(3)})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

df3 = pd.merge(df1, df2, left_on = 'avain1', right_on = 'avain2')

print(df3)
```

Äskeisissä esimerkeissä c ja d -avaimia ei tullut yhdistelmään lainkaan, koska niitä ei ollut molemmissa DataFrameissa. Oletuksena `merge` siis tekee "**inner**"-tyyppisen yhdistämisen, jossa tuloksena on vain ne avaimet, jotka ovat molemmissa DataFrameissa.

Ja koska df1:ssä oli useampia a- ja b- avaimen rivejä, tuli yhdistelmään myös useampia näitä.  

Yhdistämisen tyypin voi määrittää `how`-parametrilla, vaihtoehdot ovat:
* `ìnner` avaimet, jotka ovat molemmissa 
* `left`  vasemmassa (ensimmäisessä) DataFramessa olevat avaimet
* `right` oikeassa (jälkimmäisessä) DataFramessa olevat avaimet
* `outer` avaimet, jotka ovat jommassa kummassa DataFramessa





```python
df1 = pd.DataFrame({'avain': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'avain': ['a', 'b', 'd'],
                    'data2': range(3)})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

df3 = pd.merge(df1, df2, on = 'avain', how = 'outer')

print(df3)
```

Jos molemmissa on useita saman avaimen rivejä, `merge` tekee näistä kaikki mahdolliset "yhdistelmät", eli jos 1. DataFramessa on vaikka 4 a-avaimen riviä ja toisessa 3, on yhdistelmässä 4 x 3 = 12 a-avaimen riviä.

```python
df1 = pd.DataFrame({'avain': ['b', 'b', 'a', 'a', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'avain': ['a', 'a', 'b'],
                    'data2': range(3)})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

df3 = pd.merge(df1, df2, on = 'avain')

print(df3)
```

Entä jos yhdistettävissä DataFrameissa on samannimisiä sarakkkeita? pandas lisää päätteen (oletuksena `_x` ja `_y`) sarkakenimien perään. Omat päätteet voi määrittää parametrilla `suffixes=('_1','_2')`. 


```python
df1 = pd.DataFrame({'avain': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'avain': ['a', 'b', 'd'],
                    'data1': range(3)})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

df3 = pd.merge(df1, df2, on = 'avain')

print(df3)

print('\n------------------\n')

df4 = pd.merge(df1, df2, on = 'avain', suffixes=('_1','_2'))

print(df4)
```

Usein yhdistäminen tehdään datakentän sijaan indeksin (riviotsikon) mukaan. Tähän on parametri `left_index=True` , joka kertoo että ensimmäisessä DataFramessa yhdistämiskenttä on indeksi. Vatsaavasti `right_index=True`. 

Kerta 3:n tehtävä 5:n voisi siis aloittaa myös näin:

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/season-1718_csv.csv')

# value_counts groupby:lla antaa monitasoisen indeksin, joka saadaan unstackilla sarakkeisiin
kotona = df['FTR'].groupby(df['HomeTeam']).value_counts().unstack()
vieraissa = df['FTR'].groupby(df['AwayTeam']).value_counts().unstack()

# yhdistetään, muutetaan NaN nollaksi, ja luvut kokonaisluvuiksi (NaN tekee niistä likulukuja)
kaikki = pd.merge(kotona, vieraissa, left_index=True, right_index=True).fillna(0).astype(int)

kaikki
```

### join
Indeksin mukaan yhdistäminen voidaan tehdä myös DataFramen `join`-metodilla. Se tekee oletuksena `left`-tyyppisen yhdistämisen, mutta `how`-parametrilla voi määrittää sillekin tyypin (`left` , `right`, `outer`, `inner`).

Jos DataFrameissa on samannimisiä sarakkeita `join` ei onnistu ilman `lsuffix`- ja/tai `rsuffix`-parametreja.

`join`-metodilla voidaan yhdistää myös useampia DataFrameja kerralla
`df.join([df1, df2, df3})`

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/season-1718_csv.csv')

# value_counts groupby:lla antaa monitasoisen indeksin, joka saadaan unstackilla sarakkeisiin
kotona = df['FTR'].groupby(df['HomeTeam']).value_counts().unstack()
vieraissa = df['FTR'].groupby(df['AwayTeam']).value_counts().unstack()

kotona.join(vieraissa, rsuffix='_v')
```

### combine_first
DataFramen `combine_first`-metodia voidaan käyttää "aukkojen täyttämiseen" eli tällöin DataFramen NaN-arvot korvataan toisesta DataFramesta löytyvillä arvoilla.

```python
df1 = pd.DataFrame({'a':[4,6,7,np.nan, 5], 'b':[5, np.nan, 87,2,np.nan]})
df2 = pd.DataFrame({'a':[1,np.nan,2,6, 5, 5], 'b':[5, 1, 45,2,7, 8], 'c':[5, np.nan, 87,2,0, 9]})

print(df1)

print('\n------------------\n')

print(df2)

print('\n------------------\n')

print(df1.combine_first(df2))
```

### groupby 
Edellisellä kerralla esiteltiin `groupby`-metodi, joka ryhmittelee datan annetun kentän mukaan. Periaatteesta käytetään termiä **split-apply-combine**, eli ensin jaetaan data ryhmiin, sitten tehdään ryhmille jokin operaatio ja lopuksi yhditetään tulokset:

![split-apply-combine](https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/sac.png)  

Titanic-esimerkissä laskettiin matkustajaluokittain (Pclass) selviytyneiden määrät:

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/titanic.csv', index_col=0)

print(df['Survived'].groupby(df['Pclass']).value_counts())
```

Edellä `df['Survived']`-Series ryhmiteltiin Pclass:n mukaan, vaikka `df['Survived']`-Series ei siällä Pclass-tietoa. Tämän takia ryhmittelyperusteeksi annettiinkin `df['Pclass']`-Series, joka siis antaa ryhmittelyssä tiedon mikä indeksi kuuluu mihinkin ryhmään, ja koska `df['Survived']`-Series sisältää samat indeksit, voidaan määrät laskea ryhmittäin. 

Sama voitiin tehdä "toisin päin", eli ryhmitellään koko DataFrame ja otetaan ryhmittelystä vain Survived-sarake. Tällöin ryhmittelyperusteeksi ei tarvitse antaa `df['Pclass']` vaan riittää vain sarakkeen nimi `'Pclass'` (koska sarake löytyy df:stä).

```python
print(df.groupby('Pclass')['Survived'].value_counts())
```

`groupby`:n palauttaman objektin voi myös iteroida `for`-silmukalla, tällöin saadaan tuple (monikko) muodossa `nimi,data`:

```python
for name, data in df['Survived'].groupby(df['Pclass']):
      print("LUOKKA ",name)
      print(data.value_counts())
```

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/season-1718_csv.csv')

kotona = df['HomeTeam'].groupby(df['FTR'])

for name, data in kotona:
      print("Tulos: ",name)
      print(data.value_counts())
```

Ryhmittelyn tuottamalle `GroupBy`-objektilla on muutamia optimoituja metodeja koostelukujen laskemiseen:
* `count`	lukumäärä (ei NaN-arvoja)
* `sum`	summa
* `mean`	keskiarvo
* `median`	mediaani
* `std`, `var`	keskihajonta, varianssi
* `min`, `max`	minimi, maksimi
* `prod`	tulo
* `first`, `last`	ensimmäinen, viimeinen ei-NaN arvo

```python
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/season-1718_csv.csv')

df['HomeTeam'].groupby(df['FTR']).first()  # antaa listan (=kauden) ensimmäisten kotitappion (A), kotitasapelin (D) ja kotivoiton (H) joukkueet
```

Myös muita Serieseille määriteltyjä koostefuktioita voi käyttää: 

```python
df['FTHG'].groupby(df['FTR']).unique()  # kotijoukkueen maalimäärät kotitappioissa (A), -tasureissa (D) ja -voitoissa (H)
```

Ryhmille voi määrittää myös omia koostefunktioita käyttämällä `agg`-metodia:

```python
def vaihteluvali(x):
    return x.max()-x.min()

print(df['FTHG'].groupby(df['FTR']).agg(vaihteluvali))

print('\n------------\n')

# annetaan lista funktioista, pandasin tai itse määriettyjä
print(df['FTHG'].groupby(df['FTR']).agg([vaihteluvali, max, min]))

print('\n------------\n')

# tässä tuloksen sarakkeet nimetty itse
print(df['FTHG'].groupby(df['FTR']).agg([('vvali', vaihteluvali), ('eniten',max), ('vähiten',min)]))
```

## Ristiintaulukointi

Ristiintaulukointi eli pivotointi tarkoittaa koostamista, jossa data ryhmitellään jonkun/joidenkin kenttien mukaan riveiksi ja toiset kentät määritellään sarakkeiksi.

<a id="10"></a>


### pandas.crosstab
`pandas.crosstab`-funktiolla voidaan ristiintaulukoida frekvenssejä (lukumääriä) riveittäin/sarakkeittain.

```python

df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta3/titanic.csv', index_col=0)

pd.crosstab(df['Pclass'],  df['Survived'])
```

```python
pd.crosstab([df['Pclass'], df['Sex']], df['Survived'])
```

```python
pd.crosstab([df['Pclass'], df['Sex']], [df['Survived'], df['SibSp']])
```

Lisäämällä parametri `margins = True` saadaan myös rivi- ja sarakesummat

```python
pd.crosstab(df['Pclass'],  df['Survived'], margins = True)
```

Parametrilla `normalize = 'columns'` saadaan suhteelliset osuudet  (tässä sarakesummasta)

```python
pd.crosstab(df['Pclass'],  df['Survived'], normalize = 'columns')
```

Vastaavasti `normalize = 'index'` antaa osuudet rivisummasta ja `normalize = 'all'` kaikista arvoista.

```python
pd.crosstab(df['Pclass'],  df['Survived'], normalize = 'index')
```

Jotta suhteelliset osuudet näkyisivät prosentteina, voidaan muotoilla DataFramea `df = df.applymap("{:.1%}".format)`

```python
pd.crosstab(df['Pclass'],  df['Survived'], normalize = 'index').applymap("{:.1%}".format)
```

<a id="11"></a>

### pivot_table
`pivot_table` -funktio tarjoaa crosstabia monipuolisemman koostaa pivottaulukoita. Oletuksena `pivot_table` laskee ryhmittäin  keskiarvoja.


```python
df.pivot_table(index=['Pclass', 'Sex'])

# saman tekisi vielä df.groupby(['Pclass', 'Sex']).mean()
```

Lisäämäällä `columns`-parametrin voi määrittää myös sarakkeisiin ryhmittelyn. Myös tässä `margins=True` antaa rivi- ja sarakeyhteenvedot. 

```python
df.pivot_table( ['Age', 'Fare'], index=['Pclass', 'Sex'], columns=['Survived'], margins=True)  # tehään vain Age ja Fare-sarakkeille
```

`aggfunc` -parametrilla voidaan määrittää käytettävälksi muita koontifunktioita keskiarvon sijaan.

```python
df.pivot_table(['Age', 'Fare'], index=['Pclass', 'Sex'], columns=['Survived'], aggfunc=min)
```
