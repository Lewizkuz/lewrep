```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananalk2019/kerta6/kysely.txt', sep=';')
df = df.replace({',':'.'}, regex=True)
print(df['yleinen tyytyväisyys'].astype(float).corr(df['toimitusnopeus'].astype(float)))
print(df['yleinen tyytyväisyys'].astype(float).corr(df['hintataso'].astype(float)))
print(df['yleinen tyytyväisyys'].astype(float).corr(df['joustavuus hinnoissa'].astype(float)))
print(df['yleinen tyytyväisyys'].astype(float).corr(df['toimittajan imago'].astype(float)))
print(df['yleinen tyytyväisyys'].astype(float).corr(df['palvelu'].astype(float)))
print(df['yleinen tyytyväisyys'].astype(float).corr(df['tuotteiden laatu'].astype(float)))
#toimitusnopeus on suurin
reg=LinearRegression()
x=df['yleinen tyytyväisyys']
X=x.to_frame()
y=df['toimitusnopeus']
Y=y.to_frame()
reg.fit(X,y)
plt.scatter(x, y)
plt.plot(X,reg.predict(X))
```
