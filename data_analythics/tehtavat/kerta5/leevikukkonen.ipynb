## Teht채v채 1

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# lue arvot tiedostosta dataframeen
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta5/passatvariant.csv')
# k채ytt철철nottovuosi, matkamittari ja CO2 -histogrammit
df['kayttoonottovuosi'].plot.hist()
plt.xlabel('ik채')
plt.show()
df['matkamittarilukema'].plot.hist()
plt.xlabel('Matkalukema')
plt.show()
df['Co2'].plot.hist()
plt.xlabel('Hiilip채채st철t')
plt.show()

#k채ytt철철nottovuosi, auton pituus, massa ja CO2
sns.jointplot('kayttoonottovuosi', 'ajonKokPituus', data=df, kind='reg')
plt.show()
```

## Teht채v채 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


x = np.linspace(-1,7,1000)
y1 =  2*x +1
y2 = 0.2*x**2-3
y3 = (x+1)**0.5
fig, ax = plt.subplots()
plt.grid(b=True, which='major', axis='both', linewidth=1, linestyle='--', color = 'green')
plt.tick_params(axis='both')
plt.figtext(0.5, 0.4, 'leikkauspiste')
plt.xticks(np.arange(-1, 10, 0.5))
plt.yticks(np.arange(-5,20, 1.0))
plt.axhline(y=0, xmin=10, xmax=0, color='black')
plt.axvline(x=0, ymin=10, ymax=0, color='black')
plt.plot(x, y1, linestyle='-.')
plt.plot(x, y2)
plt.plot(x, y3, linestyle='-.')

plt.show()
```
