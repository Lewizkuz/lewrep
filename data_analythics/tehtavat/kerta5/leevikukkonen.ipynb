```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# lue arvot tiedostosta dataframeen
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta5/passatvariant.csv')
# k채ytt철철nottovuosi, matkamittari ja CO2 -histogrammit
df['kayttoonottovuosi'].plot.hist()
plt.xlabel('ik채')
plt.show()
df['matkamittarilukema'].plot.hist()
plt.xlabel('Matkalukema')
plt.show()
df['Co2'].plot.hist()
plt.xlabel('Hiilip채채st철t')
plt.show()

#k채ytt철철nottovuosi, auton pituus, massa ja CO2
sns.jointplot('kayttoonottovuosi', 'ajonKokPituus', data=df, kind='reg')
plt.show()
```
