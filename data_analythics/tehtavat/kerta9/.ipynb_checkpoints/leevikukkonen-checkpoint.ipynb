## Tehtävä 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
df = pd.read_csv('https://student.labranet.jamk.fi/~varpe/datananalk2019/kerta9/teht1.txt')
df['Läpäisy'].value_counts()
model.fit(df[['Läsnäolo', 'Harjoitukset']], df['Läpäisy'])
print('Todennäköisyydet ovat:\n')
print('20 h läsnäoloa ja 30 tehtävää:', model.predict_proba([[20,30]])[0,1])
print('24 h läsnäoloa ja 10 tehtävää:', model.predict_proba([[24,10]])[0,1])
print('15 h läsnäoloa ja 60 tehtävää:', model.predict_proba([[15,60]])[0,1])
```

## Tehtävä 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split 
df = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananalk2019/kerta9/viini.csv')
#df.fillna(0)
df.dropna(inplace=True)
x = df.drop('type', axis=1)
y = df['type']
y = y.map({'white': 1, 'red': 0})
Train, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.3)
model = LogisticRegression()
model.fit(Train, yTrain)
print('testdata: ',model.score(xTest, yTest))
```

```python
#T3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

df = pd.read_csv('http://student.labranet.jamk.fi/~varpe/datananalk2019/kerta9/teht3.txt')
df['sukupuoli'] = df['sukupuoli'].map({'nainen': 0, 'mies': 1})
model = LogisticRegression()
x=df.drop('ostaa', axis=1)
y=df['ostaa']
model.fit(x, y)
y_pred = model.predict(x)
print('Score:', model.score(x,y))
print('Conf. Matrix:', confusion_matrix(y,y_pred))

def printJust2(removed):
    x= df[['tulot','sukupuoli','naimisissa']]
    x.remove(removed)
    model.fit(x, y)
    print('Score:', model.score(x,y))
    print('Conf. Matrix:', confusion_matrix(y,y_pred))
x= df[['tulot','sukupuoli']]
printJust2('naimisissa')
printJust2('sukupuoli')
printJust2('tulot')

model.fit(x, y)
print('Score:', model.score(x,y))
print('Conf. Matrix:', confusion_matrix(y,y_pred))

x= df[['tulot','naimisissa']]
model.fit(x, y)
print('Score:', model.score(x,y))
print('Conf. Matrix:', confusion_matrix(y,y_pred))

x= df[['sukupuoli','naimisissa']]
model.fit(x, y)
print('Score:', model.score(x,y))
print('Conf. Matrix:', confusion_matrix(y,y_pred))
```
